```
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import metrics
```

```
db = pd.read_csv('diabetes.csv')
```

```
db.head()
```

```
newdb = db.iloc[:,[0,1,2,3,4,5,6,7]].values
```

```
iris_kmeans = KMeans(n_clusters=2)
newclust = iris_kmeans.fit_predict(newdb)
newclust
```

```
iris_kmeans.cluster_centers_
```

```
newdb
```

```
db['kmeans_cluster'] = iris_kmeans.labels_
```

```
db
```

```
subset = db[db.kmeans_cluster == 1]
plt.scatter(subset.Glucose, subset.DiabetesPedigreeFunction, c='red', label='cluster 1',alpha=0.3)
subset = db[db.kmeans_cluster == 0]
plt.scatter(subset.Glucose, subset.DiabetesPedigreeFunction, c='blue', label='cluster 0',alpha=0.3)

plt.legend()
```

```
for i in range(2, 10):
    k_means = KMeans(init = "k-means++", n_clusters = i, n_init = 12)
    k_means.fit_predict(newdb)
    labels = k_means.labels_
    silhouette_avg = metrics.silhouette_score(newdb, labels)
    print('silhoutte score for n_clusters = ' + str(i) + ' is ' + str(silhouette_avg))
```
