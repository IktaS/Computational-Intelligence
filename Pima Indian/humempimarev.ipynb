```
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```
db = pd.read_csv('diabetes.csv')
```

```
db.head()
```

```
newdb = db.iloc[:,[0,1,2,3,4,5,6,7]].values
```

```
iris_kmeans = KMeans(n_clusters=2)
newclust = iris_kmeans.fit_predict(newdb)
newclust
```

```
iris_kmeans.cluster_centers_
```

```
plt.scatter(newdb[newclust == 0,0], newdb[newclust == 0,1], s = 100, c = 'red', label = 'diabetes')
plt.scatter(newdb[newclust == 1,0], newdb[newclust == 1,1], s = 100, c = 'blue', label = 'tidak diabetes')

plt.scatter(iris_kmeans.cluster_centers_[:,0], iris_kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'centroid')
plt.legend()
```
