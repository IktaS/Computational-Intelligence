```
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import metrics
```

```
db = pd.read_csv('Iris.csv')
```

```
db.head()
```

```
newdb = db.iloc[:,[1,2,3,4]].values
```

```
iris_kmeans = KMeans(n_clusters=3)
newclust = iris_kmeans.fit_predict(newdb)
newclust
```

```
iris_kmeans.cluster_centers_
```

```
plt.scatter(newdb[newclust == 0,0], newdb[newclust == 0,1], s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(newdb[newclust == 1,0], newdb[newclust == 1,1], s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(newdb[newclust == 2,0], newdb[newclust == 2,1], s = 100, c = 'green', label = 'Iris-virginica')

plt.scatter(iris_kmeans.cluster_centers_[:,0], iris_kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'centroid')
plt.legend()
```

```
for i in range(2, 10):
    k_means = KMeans(init = "k-means++", n_clusters = i, n_init = 12)
    k_means.fit_predict(newdb)
    labels = k_means.labels_
    silhouette_avg = metrics.silhouette_score(newdb, labels)
    print('silhoutte score for n_clusters = ' + str(i) + ' is ' + str(silhouette_avg))
```
